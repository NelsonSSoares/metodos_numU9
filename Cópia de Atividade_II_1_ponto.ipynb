## Exercício de erros absoluto e relativos

## Dado a lista acima, faça os seguintes cálculos de erros


```python
# Erro real será dado por um número aleatório 
# (Cada execução dessa célula irá gerar um novo número)
erro_real = numero_aleatorio.uniform(18, 40)
print(erro_real)
```

### 1) Com base na média capturada com o exercício acima e o valor no erro_real, faça o cálculo do erro absoluto e erro relativo e apresente o resultado com conclusão.

Fórmula do erro absoluto e relativo:

Sendo $X$ o resultado real esperado e $X'$ o resultado aproximado obtido, logo:   

$Erro_{absoluto} = |X - X'| $   
$Erro_{relativo} = |X - X'| / |X'|$

```python
# Faça o cálculo dos erros absolutos e relativos

error_abs = abs(media_idades - erro_real)
error_rel = (error_abs / media_idades)/ media_idades

print("Erro absoluto:", error_abs)
print("Erro Relativo:", error_rel)
```
